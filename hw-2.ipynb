```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,1000)
```

```python
y1 = 5.5*np.cos(2*x)+5.5
y2 = 0.02*np.exp(x)
y3 = 0.25*x**2+0.1*np.sin(10*x)
```

```python
plt.plot(x,y1,'b')
plt.plot(x,y2,'y')
plt.plot(x,y3,'r')
plt.axis([0,2*np.pi,-1,10])
plt.xlabel('Time in ASTR/EART 119')
plt.ylabel('Measure of Awesomeness')
```
